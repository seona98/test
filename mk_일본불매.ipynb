```python
# 2단계 목록에서 href 가져오기
def mkUrlList(driver):
    url_temp = driver.current_url 
    # 메모리의 위치값에서 current_url -> 크롬 브라우저의 주소값을 넘겨줌
    for pageNo in range(45, 47, 1):
        url = url_temp + "&pageNum={pageNo}".format(pageNo=pageNo)
        print(url)
        driver.get(url)
        time.sleep(2)
        
        mkUrls_List = []
        mkUrls = driver.find_elements_by_css_selector("td .sub_list .art_tit a")
        
        for urlList in mkUrls:
            mkUrls_List.append(urlList.get_attribute("href"))
    return mkInfo(driver, mkUrls_List)
```

```python
def connectDB(dbData):
    DB_HOST = '127.0.0.1'
    DB_USER = 'root'
    DB_PASSWD = 'autoset'
    DB_NAME = 'python'
    
    conn = pymysql.connect(host=DB_HOST, user=DB_USER, password=DB_PASSWD,
                          db=DB_NAME, charset='utf8')
    curs = conn.cursor()
    sql = """insert into mknews(newsTitle,newsSubtitle,contents) values (%s, %s, %s)"""
    curs.executemany(sql,dbData)
    conn.commit()
    
    conn.close()
```

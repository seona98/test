## 뉴스 크롤링하기

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
search = '남북관계'
```

```python
url_temp = 'https://www.yna.co.kr/search/index?query={search}&ctype=A'.format(search=search)
driver = webdriver.Chrome("./chromedriver.exe")
for pageNo in range(1, 3, 1):
    url = url_temp + "&page_no={pageNo}".format(pageNo=pageNo)
    driver.get(url)
```

## URL List 가져오기
* . 은 class, # 은 id값, HTML tag는 그대로 표현됨
* driver.find_elements_by_css_selector
    * 리스트(배열), 복수 클래스가 여러 개 있을 때 한 css를 가져오려고 할 때
* driver.find_element_by_css_selector
    * 문자열, 단수, 제목같은 경우 하나만 쓰는 경우가 많음. 그럴 때.
* driver.find_elements_by_id
    * 리스트, 복수, id값이 여러 개 있을 때
* driver.find_element_by_id
    * 문자열, 단수, id 

```python
yonhapnewsUrls = driver.find_elements_by_css_selector(".cts_atclst li a")
# 값이 여러 개 같은데 하나라고 표현해서 에러남.
# aaa = driver.find_element_by_css_selector(".cts_atclst li a")
```

```python
yonhapnewsUrls_List = []
```

```python
for urlList in yonhapnewsUrls:
    yonhapnewsUrls_List.append(urlList.get_attribute("href"))
    # get_attribute - 속성값 중 href를 가져와서 print해라.
```

## 뉴스 제목 가져오기

```python
for urllist in yonhapnewsUrls_List:
        driver.get(urllist)
        # 하나씩 가져와서 뛰우기
        # newsTitle은 list
        newsTitleList = driver.find_elements_by_css_selector(".tit")
        newsTitle_temp = newsTitleList[0]
        newsTitle = newsTitle_temp.text
        print(newsTitle)
        # tit로 가져온 것 중에서 list중에서 .text하면 텍스트만 가져오는 것.
    
        newsContents = ""
        newsContentsList = driver.find_elements_by_css_selector(".article p")
        for newsContent in newsContentsList:
            newsContents = newsContents + newsContent.text
        print(newsContents)
```

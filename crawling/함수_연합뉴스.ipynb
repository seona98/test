## 함수로 만들고 DB에 등록

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time # 사이트 기동될 때 까지 2초 정도를 줌. - 셀리니움에서 get. request에서 get쓸 때
import pymysql
```

```python
# 사이트 기동
def yonhapnewsUrl(search):
        url_temp = 'https://www.yna.co.kr/search/index?query={search}&ctype=A'.format(search=search)
        driver = webdriver.Chrome("./chromedriver.exe")
        driver.get(url_temp)
        time.sleep(2)
        
        return yonhapnewsUrlList(driver)
```

```python
# 2단계 목록에서 href 가져오기
def yonhapnewsUrlList(driver):
    url_temp = driver.current_url 
    # 메모리의 위치값에서 current_url -> 크롬 브라우저의 주소값을 넘겨줌
    for pageNo in range(1, 3, 1):
        url = url_temp + "&page_no={pageNo}".format(pageNo=pageNo)
        driver.get(url)
        time.sleep(2)
        
        yonhapnewsUrls = driver.find_elements_by_css_selector(".cts_atclst li a")
        yonhapnewsUrls_List = []
        for urlList in yonhapnewsUrls:
            yonhapnewsUrls_List.append(urlList.get_attribute("href"))
    
    return yonhapnewsInfo(driver, yonhapnewsUrls_List)
```

```python
def yonhapnewsInfo(driver, yonhapnewsUrls_List):
    for urllist in yonhapnewsUrls_List:
        driver.get(urllist)
        # 하나씩 가져와서 뛰우기
        
        #newsTitle은 list
        newsTitleList = driver.find_elements_by_css_selector(".tit")
        newsTitle_temp = newsTitleList[0]
        newsTitle = newsTitle_temp.text
        print(newsTitle)
        # tit로 가져온 것 중에서 list중에서 .text하면 텍스트만 가져오는 것.
    
        newsSubtitle = ""
        newsContents = ""
        newsContentsList = driver.find_elements_by_css_selector(".article p")
        for newsContent in newsContentsList:
            newsContents = newsContents + newsContent.text
        print(newsContents)
        
        dbData = [[newsTitle, newsSubtitle, newsContents]]
        connectDB(dbData)
```

```python
def connectDB(dbData):
    DB_HOST = '127.0.0.1'
    DB_USER = 'root'
    DB_PASSWD = 'autoset'
    DB_NAME = 'python'
    
    conn = pymysql.connect(host=DB_HOST, user=DB_USER, password=DB_PASSWD,
                          db=DB_NAME, charset='utf8')
    curs = conn.cursor()
    sql = """insert into yonhapnews(newsTitle,newsSubtitle,contents) values (%s, %s, %s)"""
    curs.executemany(sql,dbData)
    conn.commit()
    
    conn.close()
```
